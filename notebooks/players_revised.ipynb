# Revising roster information 

```python
import pandas as pd
import numpy as np
import json
import os

pd.set_option('display.max_columns', None)
```

```python
columns = ['id', 'player_name', 'team_id', 'team_name', 'position', 'primary_position']
pd.DataFrame(columns=columns)
```

```python
directory = ('../raw_data/team_stats_data/')
data_list = []

for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    with open(f) as user_file:
        pdata = user_file.read()
    pdata = json.loads(pdata)
    
    team_name = pdata['name']
    team_id = pdata['id']
    
    for player in range(len(pdata["players"])):
        id = pdata['players'][player]["id"]
        player_name = f"{pdata['players'][player]['last_name']}, {pdata['players'][player]['first_name']}"
        position = pdata['players'][player]["position"]
        position = pdata['players'][player]["position"]
        primary_position = pdata['players'][player]["primary_position"]
        
        data = {'id': id,
                'player_name': player_name,
                'team_id': team_id,
                'team_name': team_name,
                'position': position,
                'primary_position': primary_position
                }
        
        data_list.append(data)
```

```python
players = pd.DataFrame(data_list)
```

```python
players = players[~players.id.duplicated(keep="first")]
players.set_index("id")
```

```python
players.to_csv("../raw_data/players.csv")
```
